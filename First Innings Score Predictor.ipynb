```python
df =df.drop(['venue','batsman','bowler','striker','non-striker'],axis=1)
```

```python
consisting_team = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
       'Mumbai Indians', 'Delhi Daredevils', 'Kings XI Punjab',
       'Royal Challengers Bangalore','Sunrisers Hyderabad']
```

```python
df=df[(df['bat_team'].isin(consisting_team)) & (df['bowl_team'].isin(consisting_team))]
```

```python
df.bat_team.unique()
```

```python
df=df[df['overs'] >= 5]
```

```python
df.head()
```

```python
from datetime import datetime
df['date'] = df['date'].apply(lambda x:datetime.strptime(x,'%Y-%m-%d'))
```

```python
df.head()
```

```python
df=df.drop('mid',axis =1)
```

```python
df.head()
```

```python
encoded_df = pd.get_dummies(df)
```

```python
pd.set_option('display.max_columns', None)
encoded_df.head()
```

```python
x_train = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year <= 2016]
x_test = encoded_df.drop(labels='total', axis=1)[encoded_df['date'].dt.year >= 2017]
```

```python
y_train = encoded_df[encoded_df['date'].dt.year <= 2016]['total'].values
y_test = encoded_df[encoded_df['date'].dt.year >= 2017]['total'].values
```

```python
x_train.drop(labels='date', axis=True, inplace=True)
x_test.drop(labels='date', axis=True, inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
y_pred = regressor.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print('MAE',mean_absolute_error(y_test,y_pred))
print('MSE',mean_squared_error(y_test,y_pred))
print('RMAE',np.sqrt(mean_absolute_error(y_test,y_pred)))
```

```python
r2_score(y_test,y_pred)
```
